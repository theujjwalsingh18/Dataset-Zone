```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
data.head() 
```

```python
data.info()
```

## Checking any null values in Dataset

```python
data[data.isnull().any(axis=1)]
```

```python
data= data.dropna().reset_index(drop=True)
```

```python
data.isnull().sum()
```

```python
## Spliting Dataset on the basis or region of Algerian 
data.loc[:122, "Region"]=0
data.loc[122:, "Region"]=1
```

```python
data.head()
```

```python
data.tail()
```

```python
data.iloc[[122]]
```

```python
## We get this raw becouse from the above the dataset ids for first region and after that the dataset is fro the second region 
## So, we have to remove the particular raw to make it clean 

data = data.drop(122).reset_index(drop=True)
```

```python
data.columns
```

```python
## As we see that their is any extra space between names of columns like ' Ws','Rain '... and so on we have to remove that also 
data.columns=data.columns.str.strip()
data.columns
```

```python
data.info()
```

```python
## Now we have to change the Datatypes of each raws into int or float accordingly 
data.head()
```

```python
data[['day','month','year','Temperature','RH','Ws']]= data[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
data[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']]= data[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']].astype(float)
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Classes'].value_counts()
```

```python
## As this columns have only two value fire and not fire but due to extra spaces it show many values in it , let's fix that 
data['Classes']=np.where(data['Classes'].str.contains('not fire'),0,1)
data['Classes'].value_counts()
```

```python
data.hist(bins=50, figsize=(15,15))
plt.show()
```

```python
## Let's look at the pie char on Classes 
perc = data['Classes'].value_counts(normalize=True)*100
classlabels=['Fire','Not Fire']
plt.figure(figsize=(5,10))
plt.pie(perc, labels=classlabels, autopct='%1.1f%%')
plt.title("Percentage of Classes")
plt.show()
```

```python
data.corr()
```

```python
plt.figure(figsize=(10,9))
cor = data.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

# Dataset is cleaned and ready to use fro futher processing i.e ML.

```python
## let's store the cleaned data into new csv file such that we have both datasets one for further analysis and one main dataset
data.to_csv('Cleaned Algerian Forest Fire Dataset.csv', index=False)
```
